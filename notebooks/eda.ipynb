# FraudWatch-ML: Exploratory Data Analysis (EDA)

## Objective
This notebook performs Exploratory Data Analysis (EDA) on the Cyber Fraud Detection dataset.

### Tasks Covered
- Data loading
- Dataset overview
- Missing value analysis
- Statistical summary
- Fraud distribution analysis
- Transaction behavior analysis
- Device and merchant analysis
- Correlation analysis
- Business insights generation


## Import Libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
plt.style.use('default')
sns.set_theme(style="whitegrid")

# Display all columns
pd.set_option('display.max_columns', None)
```

## Load Dataset

```
df = pd.read_csv("../data/cleaned/cleaned_fraud_data.csv")

# Display first 5 rows
df.head()
```

## Dataset Overview

```
print("Dataset Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

## Statistical Summary

```
df.describe(include='all')
```

## Missing Value Analysis

```
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

## Duplicate Records

```
print("Duplicate Rows:", df.duplicated().sum())
```

## Fraud vs Non-Fraud Distribution

```
plt.figure(figsize=(7,5))

sns.countplot(x='is_fraud', data=df)

plt.title("Fraud vs Non-Fraud Transactions")
plt.xlabel("Fraud Label")
plt.ylabel("Count")

plt.savefig("../visuals/fraud_distribution.png", bbox_inches='tight')

plt.show()
```

## Transaction Hour Distribution

```
plt.figure(figsize=(8,5))

sns.histplot(df['Transaction_hour'], bins=24, kde=True)

plt.title("Transaction Hour Distribution")
plt.xlabel("Hour")
plt.ylabel("Count")

plt.savefig("../visuals/transaction_hour_distribution.png", bbox_inches='tight')

plt.show()
```

## Merchant Category Distribution

```
plt.figure(figsize=(10,5))

sns.countplot(x='Merchant_category', data=df)

plt.title("Merchant Category Distribution")
plt.xticks(rotation=45)

plt.savefig("../visuals/merchant_category_distribution.png", bbox_inches='tight')

plt.show()
```

## Device Trust Score Distribution

```
plt.figure(figsize=(8,5))

sns.boxplot(y=df['Device_trust_score'])

plt.title("Device Trust Score Distribution")

plt.savefig("../visuals/device_trust_score_distribution.png", bbox_inches='tight')

plt.show()
```

## Location Mismatch Analysis

```
plt.figure(figsize=(6,4))

sns.countplot(x='Location_mismatch', data=df)

plt.title("Location Mismatch Analysis")

plt.savefig("../visuals/location_mismatch_analysis.png", bbox_inches='tight')

plt.show()
```

## Correlation Heatmap

```
plt.figure(figsize=(12,8))

correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.savefig("../visuals/correlation_heatmap.png", bbox_inches='tight')

plt.show()
```

## Top Fraudulent Merchant Categories

```
fraud_df = df[df['is_fraud'] == 1]

top_categories = fraud_df['Merchant_category'].value_counts().head(10)

plt.figure(figsize=(10,5))

top_categories.plot(kind='bar')

plt.title("Top Fraudulent Merchant Categories")
plt.xlabel("Merchant Category")
plt.ylabel("Fraud Count")

plt.savefig("../visuals/top_fraudulent_categories.png", bbox_inches='tight')

plt.show()
```

## Key Business Insights

### Insights Generated
- Fraudulent transactions can be identified using behavioral patterns.
- Device trust score may influence fraud occurrence.
- Certain merchant categories show higher fraud activity.
- Transaction timing patterns can indicate suspicious activity.
