# Cyber Fraud Detection System – Data Cleaning

## Objective
This notebook performs:
- Data loading
- Data understanding
- Data cleaning
- Missing value handling
- Duplicate removal
- Data type conversion
- Basic feature engineering
- Exporting cleaned dataset

Tech Stack:
- Python
- Pandas
- NumPy


```python
import pandas as pd
import numpy as np

# Display settings
pd.set_option('display.max_columns', None)
```

## Load Dataset

```python
# Load dataset
data = pd.read_csv('../data/raw/credit_card_fraud_raw.csv')

# Preview dataset
data.head()
```

## Dataset Overview

```python
# Shape of dataset
print("Dataset Shape:", data.shape)

# Column names
print("\nColumns:")
print(data.columns)

# Dataset information
print("\nDataset Info:")
data.info()
```

## Check Missing Values

```python
# Missing values
missing_values = data.isnull().sum()

missing_values[missing_values > 0]
```

## Handle Missing Values

```python
# Example missing value handling

# Numerical columns
num_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    data[col].fillna(data[col].median(), inplace=True)

# Categorical columns
cat_cols = data.select_dtypes(include=['object']).columns

for col in cat_cols:
    data[col].fillna(data[col].mode()[0], inplace=True)

print("Missing values handled successfully.")
```

## Check Duplicate Records

```python
# Duplicate rows
print("Duplicate Rows:", data.duplicated().sum())
```

## Remove Duplicates

```python
# Remove duplicates
data.drop_duplicates(inplace=True)

print("Duplicates removed successfully.")
print("New Shape:", data.shape)
```

## Data Type Conversion

```python
# Convert transaction_time column to datetime if present
if 'transaction_time' in data.columns:
    data['transaction_time'] = pd.to_datetime(data['transaction_time'])

print(data.dtypes)
```

## Basic Statistical Summary

```python
# Statistical summary
data.describe()
```

## Feature Engineering

```python
# Create transaction hour feature
if 'transaction_time' in data.columns:
    data['transaction_hour'] = data['transaction_time'].dt.hour

# High amount flag
if 'amount' in data.columns:
    threshold = data['amount'].quantile(0.95)
    data['high_amount_flag'] = np.where(data['amount'] > threshold, 1, 0)

print("Feature engineering completed.")
```

## Outlier Detection

```python
# Detect outliers using IQR method

if 'amount' in data.columns:
    Q1 = data['amount'].quantile(0.25)
    Q3 = data['amount'].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data['amount'] < lower_bound) | 
                    (data['amount'] > upper_bound)]

    print("Number of Outliers:", outliers.shape[0])
```

## Final Dataset Check

```python
# Final dataset preview
data.head()
```

## Export Cleaned Dataset

```python
# Save cleaned dataset
data.to_csv('cleaned_fraud_data.csv', index=False)

print("Cleaned dataset exported successfully.")
```
